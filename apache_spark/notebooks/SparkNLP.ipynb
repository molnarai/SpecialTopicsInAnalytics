```python
from sparknlp.pretrained import PretrainedPipeline

# Start Spark NLP session


# Load a pre-trained NER pipeline
ner_pipeline = PretrainedPipeline('recognize_entities_dl', lang='en')

# Annotate sample text
text = "Barack Obama was born in Hawaii and was the president of the United States."

# Run NER and print results
result = ner_pipeline.annotate(text)
print("Entities: ", list(zip(result['entities'], result['entity_labels'])))
```
